# Bessel Functions of First Order

```python
import matplotlib.pyplot as plt
plt.style.use('/Users/dialidsantiago/Git/Quant-Girl-Blog/qg_style.mplstyle')
```

```python
import numpy as np
import scipy.special as sp
```

```python
def plot_bessel_functions(pars, M):

    x = np.linspace(-M, M, 500)
    fig, ax = plt.subplots(1, 1)
    for v in pars:
        ax.plot(x, sp.jv(v, x), lw=1.5)
    my_legends = ['$J_{'+str(n)+'}(x)$' for n in pars]
    ax.legend(my_legends)
    plt.title('Bessel Functions of the 1st Kind')
    plt.show()
```

```python
plot_bessel_functions(pars=[-5, -3,  0,  3, 5], M=10);
```

```python
plot_bessel_functions(pars=[100, 120], M=200)
```

```python
def plot_modified_bessel_functions(pars, M):
    x = np.linspace(- M,M, 500)
    for v in pars:
        plt.plot(x, sp.iv(v, x))
    my_legends = ['$I_{'+str(n)+'}(x)$' for n in pars]
    plt.legend(my_legends, loc = 0)
    plt.xlabel('$x$')
    plt.ylabel('${I}_n(x)$')
    plt.title('Modified Bessel Functions of 1st Kind')
    plt.show()
```

```python
plot_modified_bessel_functions(pars=[-5, -3,  0,  3, 5], M=10)
```
