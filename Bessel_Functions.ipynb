# Bessel Functions of First Order

```python
import matplotlib.pyplot as plt
# plt.style.use("seaborn-v0_8-whitegrid") # These are optional configurations
plt.rcParams["figure.figsize"] = (12, 8)
plt.matplotlib.rc('text', usetex = True)
# plt.rcParams['axes.prop_cycle'] = cycler('color', ['1f77b4', 'ff7f0e', '2ca02c', 'd62728', '9467bd', '8c564b', 'e377c2', '7f7f7f', 'bcbd22', '17becf'])
# axes.prop_cycle: cycler('color', ['0079FF', 'FFB84C', '00DFA2', 'F266AB', 'A459D1', '5F264A', 'D4ADFC', '7f7f7f', 'B3E5BE', '97DEFF'])
%config InlineBackend.figure_format ='retina'
```

```python
import numpy as np
import scipy.special as sp


def plot_bessel_functions(pars, M):

    x = np.linspace(-M, M, 500)
    fig, ax = plt.subplots(1, 1, figsize=(12, 8))
    ax.set_prop_cycle(color=['0079ff', 'ffb84c', '00dfa2', 'f266ab', 'a459d1', '5f264a', 'd4Adfc', '7f7f7f', 'b3E5be', '97deff'])
    # fig, ax = plt.figure()
    # ax.set_prop_cycle(color=['0079FF', 'FFB84C', '00DFA2', 'F266AB', 'A459D1', '5F264A', 'D4ADFC', '7f7f7f', 'B3E5BE', '97DEFF'])
    for v in pars:
        ax.plot(x, sp.jv(v, x), lw=2.5)
    my_legends = ['$J_{'+str(n)+'}(x)$' for n in pars]
    # ax.legend(my_legends, loc = 0)
    # ax.xlabel('$x$')
    # ax.ylabel('${J}_n(x)$')
    ax.grid(axis='x', color='0.95')
    ax.grid(axis='y', color='0.95')
    # ax.grid(False)
    ax.spines[:].set_visible(False)
    plt.title('Bessel Functions of 1st Order')
    plt.show()
    #plt.savefig(’example-04-fig.pdf’)
```

```python
plot_bessel_functions(pars=[-5, -3,  0,  3, 5], M=10)
```

```python
plot_bessel_functions(pars=[100, 120], M=200)
```

```python
def plot_modified_bessel_functions(pars, M):
    x = np.linspace(- M,M, 500)
    for v in pars:
        plt.plot(x, sp.iv(v, x))
    my_legends = ['$I_{'+str(n)+'}(x)$' for n in pars]
    plt.legend(my_legends, loc = 0)
    plt.xlabel('$x$')
    plt.ylabel('${I}_n(x)$')
    plt.title('Modified Bessel Functions of 1st Order')
    plt.show()
```

```python
plot_modified_bessel_functions(pars=[-1,0, 1, 2, 3], M=3)
```

```python
plot_modified_bessel_functions(pars=[0,1,2], M=3)
```
