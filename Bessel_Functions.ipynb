# Bessel Functions of First Order

```python
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8-whitegrid") # These are optional configurations
plt.rcParams["figure.figsize"] = (12, 6)
plt.matplotlib.rc('text', usetex = True)
%config InlineBackend.figure_format ='retina'
```

```python
import numpy as np
import scipy.special as sp

def plot_bessel_functions(m, n, M):
    x = np.linspace(0, M, 500)
    pars = range(m, n)
    for v in pars:
        plt.plot(x, sp.jv(v, x))
    my_legends = ['$J_{'+str(n)+'}(x)$' for n in pars]
    plt.legend(my_legends, loc = 0)
    plt.xlabel('$x$')
    plt.ylabel('${J}_n(x)$')
    plt.title('Bessel Functions of 1st Order')
    plt.show()
    #plt.savefig(’example-04-fig.pdf’)
```

```python
plot_bessel_functions(m=0, n=3, M=10)
```

```python
plot_bessel_functions(m=100, n=102, M=500)
```
