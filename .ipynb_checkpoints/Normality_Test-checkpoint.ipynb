Notebook with a simple function for Normality tests and visual tools.

```python
from pandas_datareader import data
import matplotlib.pyplot as plt
import pandas as pd
import datetime
import numpy as np
import math
import scipy.stats as stats
import seaborn as sns
sns.set(style="whitegrid")

from scipy.stats import shapiro
from scipy.stats import anderson
import statsmodels.api as sm
from scipy.stats import skew
from scipy.stats import kurtosis
%matplotlib inline
%pylab inline
pylab.rcParams['figure.figsize'] = (16, 8)
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
def normality_test(data, col= None, visualise = True, colour = 'orchid', save = False, saveas = 'normality_test.png'):
    
    '''
    
    Function that tests a data set for normality
    
    Parameters:
    
    data: array like (if data frame, column needs to be specified)
    col (string) : column to test (if not specified, the function will take the first column)
    visualise(boolean): Displays plots if True
    save(boolean): Option to save the figures
    saveas(string): Name of the saved figure (default name is normality_test)

    
    '''
    
    if col != None:
        
        df = data.loc[:, col]
    
    else:
        
        col = data.columns.values[0]
        df = data.iloc[:,0]

    
    n = df.shape[0]
    bins = int(math.sqrt(n))   
    
    vector = df.dropna()
    mu = np.mean(vector)
    sigma = np.std(vector)
    sk = skew(vector)
    kurt = kurtosis(vector)
    
        
    if visualise == True:
        
        
        df_info = pd.DataFrame({'Metric':['Sample Size', 'Empirical Mean', 'Empirical Std', 'Empirical Skewness', 'Empirical Kurtosis'], 'Value':[n, mu, sigma, sk, kurt]})
        
        stat, p = shapiro(vector)
        df_info = df_info.append({'Metric': 'Shapiro-Wilk Statistic', 'Value' : stat}, ignore_index=True)
        df_info = df_info.append({'Metric': 'Shapiro-Wilk p-value', 'Value' : p}, ignore_index=True)

        stat, p = stats.normaltest(vector)
        df_info = df_info.append({'Metric': 'D Angostino Statistic', 'Value' : stat}, ignore_index=True)
        df_info = df_info.append({'Metric': 'D Angostino p-value', 'Value' : p}, ignore_index=True)

        stat, p =  stats.kstest(vector, 'norm')
        df_info = df_info.append({'Metric': 'Kolmogorov-Smirnov Statistic', 'Value' : stat}, ignore_index=True)
        df_info = df_info.append({'Metric': 'Kolmogorov-Smirnov p-value', 'Value' : p}, ignore_index=True)

        stat, p = sm.stats.diagnostic.normal_ad(np.array(vector))
        df_info = df_info.append({'Metric': 'Anderson-Darling Statistic', 'Value' : stat}, ignore_index=True)
        df_info = df_info.append({'Metric': 'Anderson-Darling p-value', 'Value' : p}, ignore_index=True)
        
        df_info.update(df_info[['Value']].applymap('{:,.6f}'.format))

        

        fig = plt.figure(figsize=(25,14))


        ax1 = fig.add_subplot(2, 2, 1)
        x = np.linspace(mu - 4*sigma, mu + 4*sigma, 500)
        vector.hist(bins = bins, density = True, facecolor = colour, edgecolor = colour, alpha=0.4)
        plt.plot(x, stats.norm.pdf(x, mu, sigma), "red")
        plt.axvline(mu, color='red')
        plt.xlabel(col)
        plt.ylabel('Density')
        plt.title('Comparison with Normal Density')

        ax2 = fig.add_subplot(2, 2, 2)
        sm.qqplot(vector, ax = ax2, line = 's', color = colour)
        ax2.set_title('Q-Q Plot')


        ax3 = fig.add_subplot(2, 2, 3)
        plt.plot(np.sort(vector), np.arange(1,len(vector)+1)/float(len(vector)), color = colour, label = "Population X")
        x = np.linspace(mu - 4*sigma, mu + 4*sigma, 500)
        plt.plot(x, stats.norm.cdf(x, mu, sigma), "red")
        ax3.set_title('Comparison with Normal CDF')

        ax4 = fig.add_subplot(2, 2, 4)
        font_size=14
        bbox=[0, 0, 1, 1]
        ax4.axis('off')
        
        mpl_table = ax4.table(cellText = df_info.values, bbox = bbox, colLabels=df_info.columns, loc = 'center')
        mpl_table.auto_set_font_size(False)
        mpl_table.set_fontsize(font_size)
        
        if save == True:
            
            fig.savefig(saveas)
            
        else:
            pass
        
        
    elif visualise == False:
        print('--------------------------------------------------------')    
        print('                      Sample Info                       ')        
        print('--------------------------------------------------------')
        print('Sample Size = ', n)
        print('Empirical mean = %.4f,    Empirical std = %.4f' %(mu, sigma))
        print('Empirical Skew = %.4f,    Empirical Kurtosis = %.4f' %(sk, kurt))

        print('--------------------------------------------------------')
        print('                 Testing for Normality                  ')
        print('--------------------------------------------------------')
        print('Shapiro Wilk Test')
        stat, p = shapiro(vector)
        print('Statistic = %.4f, p = %.8f' % (stat, p))

        print('--------------------------------------------------------')
        print('D Angostino Test')
        stat, p = stats.normaltest(vector)
        print('Statistic = %.4f, p = %.8f' % (stat, p))

        print('--------------------------------------------------------')      
        print('Kolmogorov-Smirnov Test')
        stat, p =  stats.kstest(vector, 'norm')
        print('Statistic = %.4f, p = %.8f' % (stat, p))

        print('--------------------------------------------------------')
        print('Anderson-Darling Test')
        stat, p = sm.stats.diagnostic.normal_ad(np.array(vector))
        print('Statistic = %.4f, p = %.8f' % (stat, p))

        print('--------------------------------------------------------')
            
    else:
        print('Visualise parameter takes values: False, True')
    
    return 1
```

```python
# Testing our function with a sample from a normal distribution

random.seed(123)
normal_data = pd.DataFrame( {'sample' : np.random.normal(0, 1, 500) })
normality_test(normal_data, visualise = True);
```

```python
# Testing our function with a sample from a t-distribution with 3 degrees of freedom

random.seed(123)
t_data = pd.DataFrame( {'sample' : np.random.standard_t(3, 500) })
normality_test(t_data, visualise = True, colour ='green');
```

## Testing Normality for Returns

Slightly modifying our previous function we get a new function which calculates the returns (standar/log) of time series and then tests for normality.

```python
def get_returns(data, kind = 'standard', normaltest = True, visualise = True, colour = 'orchid', save = False, saveas = 'normality_test.png'):
    
    '''
    
    Function that calculates the Returns (standard/log) and test for normality
    
    Parameters:
    
    frame : Data frame - must contain column ['Adj Close']
    kind (string) : Standard/log the kind of returns to be calculated
    normaltest (boolean): Tests for normalify if True
    visualise(boolean): Displays plots if True
    save(boolean) : If visualise True, we have the option to save the figure with the results
    name(string) : Name of the saved figure (default normality_test)
    
    Returns:
    frame with an extra columns containing the returns
    
    '''
    
    df = data
    
    if kind == 'standard':

        column = 'Returns'
        df.loc[:, column] = (df['Adj Close'] - df['Adj Close'].shift(1))/df['Adj Close']
    
        
    elif kind == 'log':

        column = 'Log-Returns'
        df.loc[:, column] = np.log(df['Adj Close']) - np.log(df['Adj Close'].shift(1))

    
    if normaltest == False:
        return df

        
    n = df.shape[0]
    bins = int(math.sqrt(n))   
    returns = df.loc[:, column].dropna()
    mu = np.mean(returns)
    sigma = np.std(returns)
    sk = skew(returns)
    kurt = kurtosis(returns)
    
        
    if visualise == True:
        
        
        df_info = pd.DataFrame({'Metric':['Sample Size', 'Empirical Mean', 'Empirical Std', 'Empirical Skewness', 'Empirical Kurtosis'], 'Value':[n, mu, sigma, sk, kurt]})
        
        stat, p = shapiro(returns)
        df_info = df_info.append({'Metric': 'Shapiro-Wilk Statistic', 'Value' : stat}, ignore_index=True)
        df_info = df_info.append({'Metric': 'Shapiro-Wilk p-value', 'Value' : p}, ignore_index=True)

        stat, p = stats.normaltest(returns)
        df_info = df_info.append({'Metric': 'D Angostino Statistic', 'Value' : stat}, ignore_index=True)
        df_info = df_info.append({'Metric': 'D Angostino p-value', 'Value' : p}, ignore_index=True)

        stat, p =  stats.kstest(returns, 'norm')
        df_info = df_info.append({'Metric': 'Kolmogorov-Smirnov Statistic', 'Value' : stat}, ignore_index=True)
        df_info = df_info.append({'Metric': 'Kolmogorov-Smirnov p-value', 'Value' : p}, ignore_index=True)

        stat, p = sm.stats.diagnostic.normal_ad(np.array(returns))
        df_info = df_info.append({'Metric': 'Anderson-Darling Statistic', 'Value' : stat}, ignore_index=True)
        df_info = df_info.append({'Metric': 'Anderson-Darling p-value', 'Value' : p}, ignore_index=True)
        
        df_info.update(df_info[['Value']].applymap('{:,.4f}'.format))

        
        title = 'Daily ' + column
        fig = plt.figure(figsize=(25,21))

        ax1 = fig.add_subplot(3, 2, 1)
        df['Adj Close'].plot(grid = True)
        plt.xlabel('Date')
        plt.ylabel('Adjusted Close')
        plt.title('Original Time Series')


        ax2 =fig.add_subplot(3, 2, 2)
        df.loc[:, column].plot(grid = True) 
        plt.xlabel('Date')
        plt.ylabel(title)
        plt.title(title + ' Time Series')


        ax3 = fig.add_subplot(3, 2, 3)
        x = np.linspace(mu - 4*sigma, mu + 4*sigma, 500)
        df.loc[:, column].hist(bins = bins, density = True, facecolor= colour, edgecolor = colour, alpha=0.4)
        plt.plot(x, stats.norm.pdf(x, mu, sigma), 'red')
        plt.axvline(mu, color= 'red' )
        plt.xlabel(title)
        plt.ylabel('Density')
        plt.title('Comparison with Normal Density')

        ax4 = fig.add_subplot(3, 2, 4)
        sm.qqplot(returns, ax = ax4, line = 's', color = colour)
        ax4.set_title('Q-Q Plot')


        ax5 = fig.add_subplot(3, 2, 5)
        plt.plot(np.sort(returns), np.arange(1,len(returns)+1)/float(len(returns)), color= colour, label = "Population X")
        x = np.linspace(mu - 4*sigma, mu + 4*sigma, 500)
        plt.plot(x, stats.norm.cdf(x, mu, sigma), "red")
        ax5.set_title('Comparison with Normal CDF')

        ax6 = fig.add_subplot(3, 2, 6)
        font_size=14
        bbox=[0, 0, 1, 1]
        ax6.axis('off')
        
        mpl_table = ax6.table(cellText = df_info.values, bbox = bbox, colLabels=df_info.columns, loc = 'center')
        mpl_table.auto_set_font_size(False)
        mpl_table.set_fontsize(font_size)
        
        if save == True:
            
            fig.savefig(saveas)
            
        else:
            pass
        
        
    elif visualise == False:
        print('--------------------------------------------------------')    
        print('                      Sample Info                       ')        
        print('--------------------------------------------------------')
        print('Sample Size = ', n)
        print('Empirical mean = %.4f,    Empirical std = %.4f' %(mu, sigma))
        print('Empirical Skew = %.4f,    Empirical Kurtosis = %.4f' %(sk, kurt))

        print('--------------------------------------------------------')
        print('                 Testing for Normality                  ')
        print('--------------------------------------------------------')
        print('Shapiro Wilk Test')
        stat, p = shapiro(returns)
        print('Statistic = %.4f, p= %.4f' % (stat, p))

        print('--------------------------------------------------------')
        print('D Angostino Test')
        stat, p = stats.normaltest(returns)
        print('Statistic = %.4f, p= %.4f' % (stat, p))

        print('--------------------------------------------------------')      
        print('Kolmogorov-Smirnov Test')
        stat, p =  stats.kstest(returns, 'norm')
        print('Statistic = %.4f, p= %.4f' % (stat, p))

        print('--------------------------------------------------------')
        print('Anderson-Darling Test')
        stat, p = sm.stats.diagnostic.normal_ad(np.array(returns))
        print('Statistic = %.4f, p= %.4f' % (stat, p))

        print('--------------------------------------------------------')
            
    else:
        print('Visualise parameter must take values: False, True')
    
    return df
```
