### Example Two-piece Normal: **_tpnorm_**

Let us start by creating an instance of the Two-Piece Normal which belongs to the Two-Piece-Scale family.
For this, we need to specify:
- one location parameter
- two scale parameters

````python
dist = tpnorm(loc=0.0, sigma1=1.0, sigma2=1.5)
````

The function **display_dist** plots:

    - The probability density function of our instance 'dist'
    - A histogram made with a random sample (of size 1000) from our instance 'dist

````python
display_dist(dist);
````

We can also visualise the pdf, cdf, and ppf on isolation by simply using the parameter 'show' from *dist_plot*. 
Let's take a look at the cdf.

````python
display_dist(dist, show='cdf', color='hotpink')
````

As you can see, we can choose our favorito colours to make the plots 😀

### Example Two-Piece-Shape Student t: **tpshastudent**

Now, let's create an instance from Two-Piece-Shape  which belongs to the the Two-Piece Shape family. For this, we need to specify:
- one location parameter
- one scale parameter
- two shape parameters

````python
dist = tpshastudent(loc=1.0, sigma=0.5, shape1=3.0, shape2=10.0)
````

````python
display_dist(dist);
````

### Example Double Two-Piece SinhArcSinh: **dtpsas**

Finally, let's create an instance from the Double-Two-Piece Sinh-arcsinh distribution which belongs to the Double-Two-Piece family. For this, we need to specify:
- one location parameter
- two scale parameters
- two shape parameters

````python
dist = dtpsas(loc=1.0, sigma1=5.0, sigma2=4.0, shape1=2.0, shape2=10.0)
````

````python
display_dist(dist, color='gold');
````

## 4. Methods

We provide the following methods for all supported distributions

- probability density function **pdf**
- cumulative distribution function **cdf**
- quantile function **ppf**
- random generation **random_sample**

We can evaluate **pdf, cdf, ppf** on a single value or an array type object.

### Example Two-piece normal: **tpnorm**

````python
dist = tpnorm(loc=0.0, sigma1=2.0, sigma2=1.5)
dist.pdf(0)
````

````python
dist.pdf([0.0,0.25,0.5])
````

````python
x = np.linspace(-5, 5)
y = dist.pdf(x)
plt.plot(x, y)
plt.show()
````

````python
dist.cdf([0, 1,2])
````

````python
dist.ppf(0.5)
````

````python
dist.random_sample(10)
````

## 5. Display Functions

The **twopiece** library offers three visualisation functions which dislay plots for the three families of supported distributions.

- **display_tpscale()**
- **display_tpshape()**
- **display_dtp()**

You need to pass the following parameters (as appropriate).

- ***loc***: location parameters 
- ***sigma1***: left scale parameters 
- ***sigma2***: location parameters 
- ***shape1***: shape parameters 
- ***shape2***: shape parameters
- ***show***: string 'pdf', 'cdf', 'ppf', 'random_sample', 'All' (to show all).



````python
display_tpscale(loc=0.5, sigma1=1.5, sigma2=2.5, shape=3.0, show='random_sample');
````

````python
display_tpshape(loc= 0.0, sigma=1.0, shape1=2.5, shape2=10.0, show='pdf');
````

````python
display_dtp(loc= 0.0, sigma1=1.0, sigma2=2.0, shape1=2.5, shape2=10.0, show='All');
````
